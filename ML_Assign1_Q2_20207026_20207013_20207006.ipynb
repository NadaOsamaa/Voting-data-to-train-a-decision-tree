## Calculating the euclidean distance:

```python
def euclidian_dis(a, b):    
    dim = len(str(a))
    distance = 0
    
    for i in range(16):
        distance += abs(a[i] - b[i])**2
    dis = math.sqrt(distance)
    
    return dis
```

```python
# trying the 'euclidian_dis' function:

a = [4, 7, 9, 1, 5, 16, 8, 2, 11, 3, 4, 8, 12, 7, 9, 10]
b = [10, 2, 14, 7, 9, 1, 5, 6, 8, 2, 13, 2, 4, 4, 6, 15]
euclidian_dis(a, b)
```

## gething the distance between a test point and all known data points:

```python
def KNN(X_train, X_test, y_train, y_test, k):
    
    y_hat_test = []  # we will store our output in here

    for test_point in X_test:
        distances = []

        for train_point in X_train:
            dist = euclidian_dis(test_point, train_point)
            distances.append(dist)
        
        # storing distances in a dataframe:
        df_dists = pd.DataFrame(data=distances, columns=['dist'])
        
        # sorting the distances, and only considering the k closest points:
        df_nn = df_dists.sort_values(by=['dist'], axis=0)[:k]

        # creating counter object to track the labels of k closest neighbors
        counter = Counter(y_train[df_nn.index])

        # getting the most common label of all the nearest neighbors
        prediction = counter.most_common()[0][0]
        
        # storing the prediction to the output list
        y_hat_test.append(prediction)
        
    return y_hat_test
```

## Running the model on our dataset:

```python
test = KNN(X_train, X_test, y_train, y_test, k=5)
print(test)
```
